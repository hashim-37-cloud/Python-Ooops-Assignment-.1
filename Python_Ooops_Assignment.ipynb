01. What is Object-Oriented Programming (OOP)?

Ans: Object-Oriented Programming (OOP) is a programming paradigm that uses objects and classes to organize code. OOP focuses on creating reusable objects that interact with each other, making the code more modular, flexible, and easier to maintain. The key concepts in OOP are encapsulation, inheritance, polymorphism, and abstraction.

02. What is a class in OOP?

Ans: A class is a blueprint for creating objects in OOP. It defines the properties (attributes) and behaviors (methods) that objects of that class will have. A class encapsulates data for the object and defines methods to manipulate that data.

03. What is an object in OOP?

Ans: An object is an instance of a class. It is a specific realization of the class that holds actual data and can perform actions defined by the class. Objects are created based on the class blueprint.

04. What is the difference between abstraction and encapsulation?

Ans: Abstraction is the concept of hiding complex implementation details and showing only the necessary features of an object. It focuses on what an object does, rather than how it does it.

Encapsulation is the bundling of data (attributes) and methods (functions) that operate on that data into a single unit or class. It also refers to restricting direct access to some of an object's components to prevent unauthorized modifications.

05. What are dunder methods in Python?

Ans: Dunder methods (or magic methods) are special methods in Python, which start and end with double underscores (e.g., __init__, __str__, __repr__, __add__). These methods allow the customization of basic operations (e.g., addition, string representation) for user-defined classes.

06. Explain the concept of inheritance in OOP?

Ans: Inheritance is a mechanism in OOP that allows a class to inherit attributes and methods from another class. The child class inherits the properties of the parent class, enabling code reuse and extension of functionality. It promotes a hierarchical class structure.

07. What is polymorphism in OOP?

Ans: Polymorphism allows objects of different classes to be treated as instances of the same class. It allows methods to be used in different ways, typically through method overriding or operator overloading. In Python, it is implemented through inheritance and method overriding.

08. How is encapsulation achieved in Python?

Ans: Encapsulation in Python is achieved by restricting access to an object's internal attributes and methods. This is often done using private variables and methods (prefixing with _ or __). Public methods (getters and setters) can be provided to access and modify private data.

09. What is a constructor in Python?

Ans: A constructor in Python is a special method __init__ that is automatically called when a new object is created. It is used to initialize the object's attributes.

10. What are class and static methods in Python?

Ans: Class methods are methods that operate on the class itself, not instances of the class. They are defined using the @classmethod decorator and take cls as the first parameter.

Static methods are methods that do not operate on instances or the class itself. They are defined using the @staticmethod decorator and don't take self or cls as parameters.

11. What is method overloading in Python?

Ans: Method overloading refers to defining multiple methods with the same name but different parameter lists. However, Python does not directly support method overloading, but it can be achieved by default arguments or variable-length argument lists.

12. What is method overriding in OOP?

Ans: Method overriding is when a subclass provides a specific implementation of a method that is already defined in its parent class. It allows the subclass to change the behavior of the inherited method.

13. What is a property decorator in Python?

Ans: The @property decorator is used to define a method as a getter for an attribute. It allows the method to be accessed like an attribute, providing a way to read an attribute’s value while keeping the logic of calculation or validation behind it.

14. Why is polymorphism important in OOP?

Ans: Polymorphism allows for code reuse and flexibility, enabling objects to be treated as instances of their parent class. It allows different classes to define their behavior, but still use the same interface, simplifying code maintenance and extensibility.

15. What is an abstract class in Python?

Ans: An abstract class is a class that cannot be instantiated and is meant to be subclassed. It can contain abstract methods, which are methods that must be implemented in any subclass. Python's abc module provides the necessary tools for creating abstract classes.

16. What are the advantages of OOP?

Ans: Some of the main advantages of OOP include:

* Code reuse through inheritance

* Easier to maintain and scale

* Better data security via encapsulation

* More understandable and organized code through abstraction

* Flexibility through polymorphism

17. What is the difference between a class variable and an instance variable?

Ans: Class variables are shared among all instances of a class and are defined within the class, but outside of any methods.

Instance variables are unique to each instance of a class and are typically defined within the __init__ constructor.

18. What is multiple inheritance in Python?

Ans: Multiple inheritance is a feature in Python where a class can inherit from more than one parent class. This allows the class to inherit the attributes and methods from multiple classes.

19. Explain the purpose of ‘str’ and ‘repr’ methods in Python?

__repr__ is used to define a string representation that is meant for developers and debugging, and ideally should return a string that could be used to recreate the object.

20. What is the significance of the ‘super()’ function in Python?

21. What is the significance of the del method in Python?

@classmethod defines a method that takes the class (cls) as its first argument, rather than an instance (self), and operates on class-level data.
